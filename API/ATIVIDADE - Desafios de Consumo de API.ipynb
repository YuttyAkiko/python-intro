### **Desafio 1**

Você foi contratado para criar um programa em Python que permita o envio de mensagens SMS usando a API do Twilio. Seu programa deve solicitar ao usuário que insira o número de telefone de destino e a mensagem que deseja enviar. Em seguida, ele deve usar a API do Twilio para enviar a mensagem SMS para o número especificado. Certifique-se de configurar corretamente as credenciais da sua conta do Twilio no programa para que ele possa autenticar e enviar as mensagens com sucesso. Além disso, forneça feedback ao usuário sobre o status do envio da mensagem.

```python
from twilio.rest import Client

account_sid = 'AC43be1a48f881a0c94eed86e42f1255e0'
auth_token = 'dbda1125e16bf7670abd6d36da8660ff'

# Criando um objeto client Twilio
client = Client(account_sid, auth_token)

# Solicitando ao usuário o número de telefone de destino e a mensagem
dest_phone_number = input('Digite o número de telefone de destino (formato +55(XX)XXXXX-XXXX)')
message_body = input('Digite a mensagem que você deseja enviar')

try:
    # Envia a mensagem SMS
    message = client.messages.create(
        body = message_body, 
        from_='	(334)581-9426', 
        to = dest_phone_number
        )

    # Verifica o status do envio da mensagem        
    if message.sid:
        print(f'Mensagem enviada com sucesso para {dest_phone_number}.SID: {message.sid}')
    else:
        print('O envio da mensagem falhou.')
except Exception as e:
    print(f'Ocorreu um erro ao enviar a mensagem: {str(e)}')
```
