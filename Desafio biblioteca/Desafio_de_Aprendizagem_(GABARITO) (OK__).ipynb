### Esse desafio de aprendizagem combina os conceitos que vimos até o momento, como estruturas de controle, manipulação de dicionários, listas e iterações. Ao concluir as tarefas e a reflexão, você estará aplicando diversos aspectos da programação em um cenário realista.

## Tarefas:

## 1 - Registro de Empréstimos:

### Registre os empréstimos de livros para os  alunos fictícios da escola.
## 2 -  Avaliação de Livros:

### Após os empréstimos, simule que os alunos leram os livros e atribua notas de avaliação aos livros emprestados.

## 3 - Avaliação do Atendimento da Biblioteca:

###  Registre avaliações do atendimento da biblioteca por parte dos alunos.

## 4 - Imprimir Relatório:

### Após realizar as ações anteriores, imprima um relatório que apresente as informações sobre os empréstimos, avaliações dos livros e a média das avaliações do atendimento da biblioteca.


```python
emprestimos = []
avaliacoesLivros = []
avaliacoesBiblioteca = []


def registraEmprestimo():
    print("****** Registro de Emprestimo ********")
    try:
        nome = input("Nome do usuario: ")
        livro = input("Nome do Livro: ")
        emprestimo = {"usuario": nome, "livro": livro}
        emprestimos.append(emprestimo)
        print("\n *** emprestimo cadastrado  ***\n")
        # print(emprestimos)
    except:
        print("\n *** erro ao cadastrar - tente novamente ***\n")


def avaliaLivro():
    print("****** Avaliação de Livro ********")
    try:
        nome = input("Nome do usuario: ")
        livro = input("Nome do Livro: ")
        nota = input("Digite a nota: ")
        comentario = input("Comentarios sobre o livro: ")
        avaliacao = {"usuario": nome, "livro": livro, "nota": nota, "comentario": comentario}
        avaliacoesLivros.append(avaliacao)
        print("\n *** avaliacao cadastrada ***\n")
        # print(avaliacoesLivros)
    except:
        print("\n *** erro ao cadastrar - tente novamente ***\n")

def avaliaAtendimento():
    print("****** Avaliação do Atendimento da Biblioteca ********")
    try:
        nome = input("Nome do usuario: ")
        nota = input("Digite a nota: ")
        comentario = input("Comentarios sobre o atendimento: ")
        avaliacao = {"usuario": nome, "nota": nota, "comentario": comentario}
        avaliacoesBiblioteca.append(avaliacao)
        print("\n *** avaliacao cadastrada ***\n")
        # print(avaliacoesBiblioteca)
    except:
        print("\n *** erro ao cadastrar - tente novamente ***\n")

def imprimeRelatorio():
    print("\n********* RELATÓRIO **************")
    print("********* EMPRÉSTIMOS ************")
    for item in emprestimos:
        for chave,valor in item.items():
            print(chave + "  :  " + valor)
    print("**********************************\n")
    print("******* AVALIAÇÕES LIVROS ********")
    for item in avaliacoesLivros:
        for chave,valor in item.items():
            print(chave + "  :  " + valor)
    print("**********************************\n")
    print("***** AVALIAÇÕES BIBLIOTECA ******")
    notas = 0
    numNotas = 0
    for item in avaliacoesBiblioteca:
        for chave,valor in item.items():
            print(chave + "  :  " + valor)
            if chave=="nota":
                notas += int(valor)
                numNotas += 1
    if(notas !=0 ):
      media = int(notas) / int(numNotas)
      conceito = ""
      if(media >= 7):
        conceito = "ÓTIMO"
      elif(media < 7) and (media >= 5):
        conceito = "BOM"
      else:
        conceito = "PRECISA MELHORAR!!"
        print(f"Nota média das avaliações é: {media} - {conceito}")
    else:
        print(f"Nota média das avaliações é 0 - PRECISA MELHORAR!!")

    print("**********************************\n")


def mostraMenu():
    print('''***** Menu de opções - Digite a opção e tecle ENTER: ****

        1. Registro de Empréstimo de Livro
        2. Avaliar Livro
        3. Avaliar Atendimento da Biblioteca
        4. Imprimir Relatorio
        5. Finalizar Programa''')

def analisaOpcao(opcao):
    if (opcao == "5"):
        print("Programa Finalizado")
    elif (opcao=="1") or (opcao=="2") or (opcao=="3") or (opcao=="4"):
        executaOpcao(opcao)
    else:
        print('\n*** opcao invalida ***\n')

def executaOpcao(opcao):
    if(opcao == "1"): registraEmprestimo()
    if(opcao == "2"): avaliaLivro()
    if(opcao == "3"): avaliaAtendimento()
    if(opcao == "4"): imprimeRelatorio()

def programaPrincipal():
    opcao = 0
    while(opcao != "5"):
        mostraMenu()
        opcao = input('''
        >>>>>>>>>>> Digite a opção desejada:  ''')
        analisaOpcao(opcao)

programaPrincipal()




```
